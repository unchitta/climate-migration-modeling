```python
import numpy as np
import pandas as pd
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler()
```

```python
countries = pd.read_csv("data/countries.csv")
countries = countries.set_index('Unnamed: 0')
```

```python
yes = scaler.fit_transform(countries[['GDP(2015-2024)','Employment Rate', 'Pop Density', "Population", 'Refugee Population', "Number of CH sights"]])
```

```python
newdataframe = pd.DataFrame(yes, index = countries.index, columns = ["normGDP",'normEmp', 'normPopDens', "normPop", 'normRef', "normCH"])
```

```python
result = pd.merge(countries, newdataframe, on="Unnamed: 0")
result
```

So Now all of our data is normalized by minmax, we create our function

```python
countries["Viability"] = 0
```

```python
a = .3
b = .3
c = .2
d = .2

def viability(country):
    return a*(country["normGDP"]) + b*(1-country['normPopDens'])*(country['normPop']) + c*(country['normEmp']) - d*(country['Vulnerability score to climate change(ND-Gain index) in 2017'])
```

```python
for i,j in countries.iterrows():
    countries.loc[i , "Viability"]  = viability(result.loc[i])
```
